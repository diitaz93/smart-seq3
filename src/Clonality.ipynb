## Import TCR datasets

```python
def read_dataframe(in_file):
    file_type = in_file.split('.')[-1]
    if file_type == 'tsv':
        return pd.read_csv(in_file,sep='\t',index_col=0)
    elif file_type == 'xlsx':
        return pd.read_excel(in_file,index_col=0)
    elif file_type == 'csv':
        return pd.read_csv(in_file,sep=',',index_col=0)
    else:
        raise NameError("Invalid input format. Has to be either .tsv, .csv or .xlsx.")
```

```python
DF_1 = read_dataframe('../data/05_SS3_collected_TCRs/placa_pepita_A/placa_pepita_A.csv')
DF_2 = read_dataframe('../data/05_SS3_collected_TCRs/placa_pepita_B/placa_pepita_B.csv')
```

## Merge
